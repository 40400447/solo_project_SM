```python
import numpy as np
import matplotlib.pyplot as plt

distance = np.linspace(0, 30, 200)

p_correct = np.minimum(0.5, distance*3 / 80)

plt.figure()
plt.plot(distance, p_correct, linewidth=3)
plt.xlabel("Distance of penalty spot (yards)", fontsize=20)
plt.ylabel("$P$(correct guess)",fontsize=20)


plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.tight_layout()
plt.savefig("guess_probability.pdf")

plt.show()
```
