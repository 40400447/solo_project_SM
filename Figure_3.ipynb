```python
import numpy as np
import matplotlib.pyplot as plt

BALL_SPEED = 110       # ft/s
REACTION_TIME = 0.25   # s
DIVE_SPEED = 16        # ft/s
ARM_REACH = 2          # ft

distance = np.linspace(0, 30, 200)

travel_time = distance * 3 / BALL_SPEED   # 1 yard = 3 feet
move_time = np.maximum(0, travel_time - REACTION_TIME)

reach = ARM_REACH + DIVE_SPEED * move_time

plt.figure(figsize=(8,6))
plt.plot(distance, reach, linewidth=3, label="Reachable distance")

plt.xlabel("Distance of penalty spot (yards)", fontsize=24)
plt.ylabel("Reachable distance (feet)", fontsize=24)

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.tight_layout()
plt.savefig("reach_with_point.pdf")
plt.show()
```
