```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 30, 400)
y = x/37

reaction = 0.25

points = [12]

plt.figure(figsize=(8,6))

plt.plot(x, y, linewidth=4, label='Ball travel time', zorder=1)

plt.axhline(y=reaction,
            linewidth=4,
            linestyle='--',
            color='orange',
            label="Goalkeeper reaction time",
            zorder=2)

for d in points:

    t = d / 37

    plt.scatter(d, t,
                color='red',
                s=250,
                zorder=3)

    plt.text(d-2.5, t+0.04,
             f"{d}yds\n{t:.3f}s",
             fontsize=16,
             weight='bold',
             zorder=4)

plt.xlabel("Distance of penalty spot (yards)", fontsize=24)
plt.ylabel("Time to goal (seconds)", fontsize=24)

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.legend(fontsize=20)

plt.tight_layout()

plt.savefig("travel_time.pdf")
plt.show()

```
